```python
//значение коээфициента ближе к 1 означает наибольшее влияние на цену, но как не выводить столбец SalePrice?
corr=data.corr()["SalePrice"]
corr[np.argsort(corr, axis=0)[::-1]]
```

```python
corr = data.corr()
print ("*Признак, наиболее коррелирующий с ценой:*")
print (corr['SalePrice'].sort_values(ascending=False)[:2],'\n')
```
